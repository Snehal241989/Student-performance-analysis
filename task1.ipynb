```python
# Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load Dataset
df = pd.read_csv("student_details.csv")  

# Initial Data Preview
print(df.head())
print(df.info())

# Create Target Variable
# If average score >= 50, then Pass (1), else Fail (0)
df['Pass'] = (df['Average_Score'] >= 50).astype(int)
print(df['Pass'].value_counts())

# Visualizing Score Distributions
sns.set(style="whitegrid")
plt.figure(figsize=(16, 10))

plt.subplot(2, 2, 1)
sns.histplot(df['Math score'], kde=True, color='skyblue')
plt.title('Math Score Distribution')

plt.subplot(2, 2, 2)
sns.histplot(df['Reading score'], kde=True, color='purple')
plt.title('Reading Score Distribution')

plt.subplot(2, 2, 3)
sns.histplot(df['Writing score'], kde=True, color='red')
plt.title('Writing Score Distribution')

plt.subplot(2, 2, 4)
sns.countplot(data=df, x='Pass', palette='pastel')
plt.title('Pass/Fail Distribution')
plt.xticks([0, 1], ['Fail', 'Pass'])

plt.tight_layout()
plt.show()

#  Categorical vs Pass Analysis
categorical_cols = ['Gender', 'race/ethnicity', 'Parental level of education', 'Lunch', 'Test preparation course']
plt.figure(figsize=(18, 12))

for i, col in enumerate(categorical_cols, 1):
    plt.subplot(3, 2, i)
    sns.countplot(data=df, x=col, hue='Pass', palette='Set2')
    plt.title(f'{col} vs Pass/Fail')
    plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```
